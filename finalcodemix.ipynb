```python
# Simple Hinglish-to-Hindi Translator
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

# Load model (automatically uses GPU if available)
model = AutoModelForSeq2SeqLM.from_pretrained(
    "ai4bharat/IndicTrans2-indic-indic-1B",
    trust_remote_code=True
).to("cuda" if torch.cuda.is_available() else "cpu")

tokenizer = AutoTokenizer.from_pretrained(
    "ai4bharat/IndicTrans2-indic-indic-1B",
    trust_remote_code=True
)

def hinglish_to_hindi(text):
    """Converts Hinglish (Hindi-English mix) to pure Hindi"""
    input_text = f"hin_Deva hin_Deva {text}"  # Source and target both Hindi
    inputs = tokenizer(input_text, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example usage
print(hinglish_to_hindi("Aaj mera mood off hai"))       # Output: आज मेरा मूड खराब है
print(hinglish_to_hindi("Meeting cancel kar do"))       # Output: मीटिंग रद्द कर दो
print(hinglish_to_hindi("Tumhara response late hai"))   # Output: तुम्हारा जवाब देर से आया है
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

# Load model
model = AutoModelForSeq2SeqLM.from_pretrained(
    "ai4bharat/IndicTrans2-indic-indic-1B",
    trust_remote_code=True
).to("cuda" if torch.cuda.is_available() else "cpu")

tokenizer = AutoTokenizer.from_pretrained(
    "ai4bharat/IndicTrans2-indic-indic-1B",
    trust_remote_code=True
)

def bengalish_to_bengali(text):
    """
    Approximate Bengalish-to-Bengali translation using eng_Latn as source
    """
    input_text = f"eng_Latn ben_Beng {text}"
    inputs = tokenizer(input_text, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Try some examples
print(bengalish_to_bengali("Ami school jabo kalke"))       # Expected: আমি স্কুল যাব কালকে
print(bengalish_to_bengali("Tumi bhalo acho"))              # Expected: তুমি ভালো আছো
print(bengalish_to_bengali("Porikkha khub tough chhilo"))   # Expected: পরীক্ষা খুব কঠিন ছিল
```

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import torch

# Model name for transliteration
translit_model_name = "shadabtanjeed/mbart-banglish-to-bengali-transliteration"

# Load tokenizer and model
tokenizer_translit = MBart50TokenizerFast.from_pretrained(translit_model_name)
model_translit = MBartForConditionalGeneration.from_pretrained(translit_model_name)
device = "cuda" if torch.cuda.is_available() else "cpu"
model_translit = model_translit.to(device)

# Set tokenizer to correct language codes if needed
# For mBART-based transliteration, sometimes you set tokenizer.src_lang:
tokenizer_translit.src_lang = "en_XX"  # or appropriate code used during fine-tuning
```

```python
def transliterate_bengalish_to_bengali_script(text: str, max_length: int = 128) -> str:
    """
    Given a Romanized Bengali (Bengalish) string, return Bengali script via the transliteration model.
    """
    # Tokenize
    inputs = tokenizer_translit(text, return_tensors="pt", truncation=True, max_length=max_length).to(device)
    # Generate output IDs
    generated_ids = model_translit.generate(
        **inputs,
        max_length=max_length,
        num_beams=4,
        early_stopping=True,
    )
    # Decode
    transliterated = tokenizer_translit.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return transliterated

# Example usage:
bengalish_example = "tumi ki se'i chota barita dekhate pa'o?"
bengali_script = transliterate_bengalish_to_bengali_script(bengalish_example)
print("Bengali script:", bengali_script)

```

```python
from transformers import MarianMTModel, MarianTokenizer
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

# Load tokenizer and model
tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-bn-en")
model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-bn-en").to(device)

def translate_bengali_to_english_opus(text: str) -> str:
    """
    Translate Bengali script text to English using Helsinki-NLP/opus-mt-bn-en.
    """
    # Tokenize
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
    # Generate
    outputs = model.generate(**inputs, max_length=128, num_beams=4, early_stopping=True)
    # Decode
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example test
bengali_input = "আমার নাম রাহুল এবং আমি কলকাতায় থাকি।"
english_output = translate_bengali_to_english_opus(bengali_input)
print("Input:", bengali_input)
print("Output:", english_output)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

# Load transliteration model
translit_model_name = "gowtham58/T_TL"
tokenizer_tl = AutoTokenizer.from_pretrained(translit_model_name)
model_tl = AutoModelForSeq2SeqLM.from_pretrained(translit_model_name).to(device)

def tanglish_to_tamil(text: str) -> str:
    """
    Converts Tanglish (Romanized Tamil) to Tamil script using the Hugging Face model.
    """
    # Directly feed the Roman text; the model knows to transliterate
    inputs = tokenizer_tl(text, return_tensors="pt", truncation=True, padding=True).to(device)
    outputs = model_tl.generate(**inputs, max_length=128, num_beams=4, early_stopping=True)
    tamil = tokenizer_tl.decode(outputs[0], skip_special_tokens=True)
    return tamil

# Test transliteration
examples = [
    "Inime ipdi nadakkakoodadhu",

]

for ex in examples:
    tamil_script = tanglish_to_tamil(ex)
    print(f"Tanglish: {ex}\nTamil:    {tamil_script}\n")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from IndicTransToolkit.processor import IndicProcessor

# Device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load model & tokenizer
model_name = "ai4bharat/indictrans2-en-indic-1B"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    torch_dtype=torch.float16,
    attn_implementation="flash_attention_2"
).to(device)

# Preprocessor
ip = IndicProcessor(inference=True)

# Input Onglish
raw = "Tume kahaku bhala kahiba?"
src_lang = "eng_Latn"

# Translation function (fixed)
def translate(src_lang, tgt_lang, text):
    # No manual tag added here
    batch = ip.preprocess_batch([text], src_lang=src_lang, tgt_lang=tgt_lang)
    inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True).to(device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=128, num_beams=5)
    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    cleaned = ip.postprocess_batch(decoded, lang=tgt_lang)
    return cleaned[0]

# Generate translations
pure_odia = translate(src_lang, "ory_Orya", raw)
print("✅ Pure Odia:", pure_odia)

pure_eng = translate(src_lang, "eng_Latn", raw)
print("✅ Pure English:", pure_eng)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from IndicTransToolkit.processor import IndicProcessor

# ✅ Transliteration import
from indic_transliteration.sanscript import transliterate, SCHEMES

# Device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load IndicTrans2 (Indic-to-English model)
model_name = "ai4bharat/indictrans2-indic-en-1B"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    torch_dtype=torch.float16
).to(device)

ip = IndicProcessor(inference=True)

# Onglish input
raw = "Tume kebe asiba"

# ✅ Transliterate from ITRANS-style to native Odia script
odia_native = transliterate(raw, "itrans", "oriya")
print("🔄 Native Odia:", odia_native)

# Translation function
def translate_odia_to_eng(text_native):
    batch = ip.preprocess_batch([text_native], src_lang="ory_Orya", tgt_lang="eng_Latn")
    inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True).to(device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=128, num_beams=5)
    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    cleaned = ip.postprocess_batch(decoded, lang="eng_Latn")
    return cleaned[0]

# Translation output
pure_odia = odia_native
pure_eng = translate_odia_to_eng(odia_native)

print("✅ Pure Odia:   ", pure_odia)
print("✅ Pure English:", pure_eng)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from IndicTransToolkit.processor import IndicProcessor

# ✅ Set device (GPU if available)
device = "cuda" if torch.cuda.is_available() else "cpu"

# ✅ Load model and tokenizer
model_name = "ai4bharat/indictrans2-en-indic-1B"

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)

model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    attn_implementation="flash_attention_2" if device == "cuda" else "eager"
).to(device)

# ✅ Load the Indic processor
ip = IndicProcessor(inference=True)

# ✅ Translation function
def translate(src_lang, tgt_lang, text):
    batch = ip.preprocess_batch([text], src_lang=src_lang, tgt_lang=tgt_lang)
    inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True).to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=128, num_beams=5)
    decoded = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    cleaned = ip.postprocess_batch(decoded, lang=tgt_lang)
    return cleaned[0]

# ✅ Roman Urdu input
raw_text = "Kya tumne homework complete kiya"

# ✅ Translate to Pure Urdu (Arabic script)
pure_urdu = translate("eng_Latn", "urd_Arab", raw_text)

# ✅ Output
print("✅ Pure Urdu:", pure_urdu)
```
